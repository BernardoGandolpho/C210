## K-Means

### Criação de dados

```python
from sklearn.datasets import make_blobs
```

```python
data, _ = make_blobs(n_samples=300, n_features=2, centers=3, cluster_std=0.5, shuffle=True)
```

Por ser um exemplo de aprendizado não supervisionado, não precisamos do retorno de com as respostas, e por isso marcamos com '_'

### Análise Exploratória

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.dpi"] = 150

plt.scatter(data[:, 0], data[:, 1], marker="o", edgecolors="black")

plt.show()
```

### Criação do modelo K-Means

```python
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters=3)
modelo.fit(data)
```

### Predições (Resposta do modelo)

```python
resposta = modelo.predict(data)
```

#### Visualizando a resposta

```python
grupo_1 = data[resposta == 0]
grupo_2 = data[resposta == 1]
grupo_3 = data[resposta == 2]

centros = modelo.cluster_centers_
```

```python
plt.scatter(grupo_1[:, 0], grupo_1[:, 1], marker="o", c="blue", label="Grupo 1", edgecolor="black")
plt.scatter(grupo_2[:, 0], grupo_2[:, 1], marker="s", c="green", label="Grupo 2", edgecolor="black")
plt.scatter(grupo_3[:, 0], grupo_3[:, 1], marker="v", c="red", label="Grupo 3", edgecolor="black")

plt.scatter(centros[:, 0], centros[:, 1], marker="*", c="yellow", label="Centros")

plt.legend(scatterpoints=True)
plt.show()
```

#### Métricas de avaliação

```python
from sklearn.metrics import calinski_harabasz_score, silhouette_score

sh = silhouette_score(data, modelo.labels_)
ch = calinski_harabasz_score(data, modelo.labels_)
```

```python
print("SH:", sh)
print("CH:", ch)
```
